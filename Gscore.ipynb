# G score function :
---
the gscore function allow us to perform our models taken into account two factors in one time the time token and the accuracy, in our case we used the f_beta_score and the Tw unit is by millisecondes :

image.png

we decide that alpha + beta = 1

```python
def Gscore(alpha,beta,fscores,timeToken):
  return (alpha*sum(timeToken))/(beta*sum(fscores))

Gscore(0.9,0.1,[0.1,0.5,0.8,0.4,0.1,0.5,0.93],[0.1,0.5,0.8,0.9,0.6,0.3,0.47,0.7])
```
